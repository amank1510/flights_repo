### Incremental Data Ingestion

```python
dbutils.widgets.text("data_source","")
```

```python
data_source=dbutils.widgets.get("data_source")
data_source
```

```python
df=(
    spark.readStream
    .format("cloudFiles")
    .option("cloudFiles.format","csv")
    .option("cloudFiles.schemaLocation",f"/Volumes/flights/bronze/bronze_volume/{data_source}/checkpoint")
    .option("cloudFiles.schemaEvolutionMode","rescue")
    .load(f"/Volumes/flights/raw/raw_volume/{data_source}/*.csv")
)
```

```python
df.writeStream\
    .format("delta")\
    .outputMode("append")\
    .trigger(once=True)\
    .option("checkPointLocation",f"/Volumes/flights/bronze/bronze_volume/{data_source}/checkpoint")\
    .option("path",f"/Volumes/flights/bronze/bronze_volume/{data_source}/data")\
    .start()
```
