```python
dbutils.widgets.text("catalog","")
dbutils.widgets.text("schema","")
```

```python
catalog_name=dbutils.widgets.get("catalog")
schema_name=dbutils.widgets.get("schema")
```

```python
spark.sql(f"CREATE CATALOG IF NOT EXISTS {catalog_name}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog_name}.{schema_name}")
```

```python
def createExternalVolume(catalog_name,schema_name,volume_name,base_path):
    if(schema_name=="raw"):
        spark.sql(f"""
                CREATE EXTERNAL VOLUME IF NOT EXISTS {catalog_name}.{schema_name}.{schema_name}_volume
                LOCATION '{base_path}'
                COMMENT "Volume for raw data"
                """)
    else:
        spark.sql(f"""
                  CREATE VOLUME IF NOT EXISTS {catalog_name}.{schema_name}.{schema_name}_volume
                  COMMENT 'Volume for {schema_name}'
                  """)
```

```python
createExternalVolume(catalog_name,schema_name,volume_name,base_path)
```
